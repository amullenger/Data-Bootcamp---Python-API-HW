```python
# Dependencies and Setup
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
import requests
import time

# Import API key
import api_keys
ow_key = api_keys.api_key

# Incorporated citipy to determine city based on latitude and longitude
from citipy import citipy

# Output File (CSV)
output_data_file = "output_data/cities.csv"

# Range of latitudes and longitudes
lat_range = (-90, 90)
lng_range = (-180, 180)
ow_key
```

## Generate Cities List

```python
# List for holding lat_lngs and cities
lat_lngs = []
cities = []

# Create a set of random lat and lng combinations
lats = np.random.uniform(low=-90.000, high=90.000, size=1500)
lngs = np.random.uniform(low=-180.000, high=180.000, size=1500)
lat_lngs = zip(lats, lngs)

# Identify nearest city for each lat, lng combination
for lat_lng in lat_lngs:
    city = citipy.nearest_city(lat_lng[0], lat_lng[1]).city_name
    
    # If the city is unique, then add it to a our cities list
    if city not in cities:
        cities.append(city)

# Print the city count to confirm sufficient count
len(cities)
```

```python
#cities
```

### Perform API Calls
* Perform a weather check on each city using a series of successive API calls.
* Include a print log of each city as it'sbeing processed (with the city number and city name).


```python
# Create dictionary for storing desired weather values from API calls
weather_dict = {
    "City": [],
    "Country": [],
    "Temperature": [],
    "Humidity": [],
    "Cloudiness": [],
    "Wind Speed": [],
    "Date": []
}

# Create list for storing latitudes for all cities for which we are able to sucessfully pull data
city_lats = []

# Create a formatted URL to which we will add each city name as we make API calls
base_url = "http://api.openweathermap.org/data/2.5/weather?"
units = "imperial"
formatted_url = f"{base_url}appid={ow_key}&units={units}&q="

# Loop through city names to make API calls. If the request is not successful, move on to the 
# next city, otherwise append the necessary data points to weather dictionary
for city in cities:
    query_url = f"{formatted_url}{city}"    
    response = requests.get(query_url)
    if response.status_code == 200:
        city_json = response.json()
        
        city_name = city_json['name']
        weather_dict['City'].append(city_name)

        city_country = city_json['sys']['country']
        weather_dict['Country'].append(city_country)

        city_temp = city_json['main']['temp_max']
        weather_dict['Temperature'].append(city_temp)

        city_hum = city_json['main']['humidity']
        weather_dict["Humidity"].append(city_hum)

        city_cloud = city_json['clouds']['all']
        weather_dict["Cloudiness"].append(city_cloud)

        city_wind = city_json['wind']['speed']
        weather_dict['Wind Speed'].append(city_wind)
        
        city_dt_code = city_json['dt']
        weather_dict['Date'].append(city_dt_code)
        
        lat = city_json['coord']['lat']
        city_lats.append(lat)
    else:
        continue
print("success")
```

### Convert Raw Data to DataFrame
* Export the city data into a .csv.
* Display the DataFrame

```python
# Generate dataframe from weather dictionary
weather_df = pd.DataFrame(weather_dict)

# Save dataframe to csv file
weather_df.to_csv("weather_data.csv", header = True)

weather_df
```

```python
# Check that there are at least 500 data points in the dataframe. ALso check that this number
# matches the number of latitudes we stored
print(weather_df.count())
print(len(city_lats))
```

#### Latitude vs. Temperature Plot

```python
# Store today's date as a formatted string for use in plot titles
struct_time_obj = time.localtime(time.time())
today_month = struct_time_obj[1]
today_day = struct_time_obj[2]
today_yr = struct_time_obj[0]
f_date = f"{today_month}/{today_day}/{today_yr}"
```

```python
# Get x and y coordinates from city latitudes list and weather_df. X data will be reused for 
# subsequent plots
x_data = city_lats
y_temp_data = weather_df['Temperature']

# Format scatter plot
plt.scatter(x_data, y_temp_data, alpha = .8)
plt.xlabel("Latitude")
plt.ylabel("Temperature(F)")
plt.title(f"City Temperature vs. Latitude - {f_date}")
plt.grid(which = "both", c = "gray", alpha = .7)

# Save plot as .png file
plt.savefig("lat_vs_temp_scatter.png", edgecolor = "black")

plt.show()
```

#### Latitude vs. Humidity Plot

```python
# Get y coordinates from weather_df, x coordinates stored above
y_hum_data = weather_df["Humidity"]

# Format scatter plot
plt.scatter(x_data, y_hum_data, alpha = .8)
plt.xlabel("Latitude")
plt.ylabel("Humidity(%)")
plt.title(f"City Latitude vs. Humidity(%) - {f_date}")
plt.grid(which = "both", c = "gray", alpha = .7)

# Save plot as .png file
plt.savefig("lat_vs_hum_scatter.png", edgecolor = "black")

plt.show()
```

#### Latitude vs. Cloudiness Plot

```python
# Get y coordinates from weather_df, x coordinates stored above
y_data_cloud = weather_df['Cloudiness']

# Format scatter plot
plt.scatter(city_lats, y_data_cloud, alpha = .8)
plt.xlabel("Latitude")
plt.ylabel("Cloudiness(%)")
plt.title(f"City Latitudes vs. Cloudiness(%) - {f_date}")
plt.grid(which = "both", c = "gray", alpha = .7)

# Save plot as .png file
plt.savefig("lat_vs_cloudiness_scatter.png", edgecolor = "black")

plt.show()
```

#### Latitude vs. Wind Speed Plot

```python
# Get y coordinates from weather_df, x coordinates stored above
y_data_wind = weather_df["Wind Speed"]

# Format scatter plot
plt.scatter(city_lats, y_data_wind, alpha = .8)
plt.xlabel("Latitude")
plt.ylabel("Windiness(%)")
plt.title(f"City Latitude vs. Wind Speed(mph) - {f_date}")
plt.grid(which = "both", c = "gray", alpha = .7)

# Save plot as .png file
plt.savefig("lat_vs_windspeed_scatter.png", edgecolor = "black")

plt.show()
```
